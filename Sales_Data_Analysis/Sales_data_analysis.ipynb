## Recommendations:

### 1. Identifying Staffing shortages:

To determine which facilities have the lowest staffing per resident, potentially indicating workforce shortages and a need for contract staffing solutions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pbj_df['Total_Staffing_Hours'] = pbj_df['Hrs_RN'] + pbj_df['Hrs_LPN'] + pbj_df['Hrs_CNA']
```

```python
# Group by facility and sum total staffing hours
staffing_summary = pbj_df.groupby('PROVNAME', as_index=False).agg({
    'Total_Staffing_Hours': 'sum',
    'MDScensus': 'sum'
})

# Calculate staffing per resident
staffing_summary['Staffing_per_Resident'] = staffing_summary['Total_Staffing_Hours'] / staffing_summary['MDScensus']

# Drop NaN values and facilities with zero staffing
staffing_summary = staffing_summary.dropna().query("Total_Staffing_Hours > 0")

# Convert Staffing_per_Resident (hours) into HH:MM:SS format using pandas timedelta
staffing_summary['Staffing_per_Resident_Time'] = pd.to_timedelta(staffing_summary['Staffing_per_Resident'], unit='h')

# Sort and select top 20 facilities with lowest staffing per resident
staffing_summary = staffing_summary.sort_values(by='Staffing_per_Resident', ascending=True).head(20)
```

```python
# Extract minutes from Staffing_per_Resident_Time
staffing_summary['Staffing_per_Resident_Minutes'] = staffing_summary['Staffing_per_Resident_Time'].dt.total_seconds() / 60  # Convert seconds to minutes

# Define colors for different staffing categories
colors = ['#005F73' if i < 5 else 'grey' for i in range(len(staffing_summary))]

# Plot
fig, ax = plt.subplots(figsize=(14, 8))

# Set a custom background color
fig.patch.set_facecolor('#fffbf7')  # Figure background color
ax.set_facecolor('#fffbf7')        # Axis background color

# Create horizontal bar plot using minutes
sns.barplot(data=staffing_summary,
            x='Staffing_per_Resident_Minutes',
            y='PROVNAME',
            palette=colors,
            ax=ax)

# Customize title and axis labels
ax.set_title("Top 20 Understaffed Nursing Facilities (Lowest Staffing Time per Resident in 2024Q2)",
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Average Staffing Time per Resident (Minutes)", fontsize=14, labelpad=10)
ax.set_ylabel("Facility Name", fontsize=14, labelpad=10)

# Rotate Y-axis labels for better readability
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right", fontsize=12)

# Add horizontal gridlines
ax.yaxis.grid(True, color="#D3D3D3", linestyle="--", linewidth=0.5, alpha=0.7)
ax.xaxis.grid(False)

# Simplify spines (remove top and right borders)
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

# Annotate bars with staffing per resident value in minutes
for index, value in enumerate(staffing_summary['Staffing_per_Resident_Minutes']):
    ax.text(value + 1, index, f"{int(value)} min", va='center', fontsize=10, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
staffing_summary
```

What Does This Mean?

1. There is a Critical Need for More Staffing.

- The CMS standard is 4.1 hours per resident per day, but none of these facilities even reach 1 hour.

- These nursing homes are likely struggling with severe staffing shortages.

2. High-Risk Facilities Need Immediate Attention.

- Facilities with staffing per resident under 1 hour may be failing to meet basic care standards.

- This can lead to poor health outcomes, safety concerns, and legal risks.

3. Clipboard Health Should Target These Facilities.

- Facilities in this dataset are prime candidates for contract staffing solutions.

- Clipboard Health can offer RN, LPN, and CNA placements to help these nursing homes meet CMS requirements.

### 2. Analyze Contractor vs. Employee Staffing

To determine which facilities rely heavily on contractors instead of full-time employees. High contractor usage may indicate:

- Retention issues (facilities struggle to retain full-time staff).
- Budget constraints (facilities may prefer temporary staffing).
- Operational inefficiencies (fluctuating workforce may affect care quality).

```python
# Group by facility and sum contractor and total staffing hours
contractor_summary = pbj_df.groupby('PROVNAME', as_index=False).agg({
    'Hrs_RN_ctr': 'sum',
    'Hrs_LPN_ctr': 'sum',
    'Hrs_CNA_ctr': 'sum',
    'Total_Staffing_Hours': 'sum'  # Already calculated earlier
})

# Calculate total contractor hours
contractor_summary['Total_Contractor_Hours'] = contractor_summary['Hrs_RN_ctr'] + contractor_summary['Hrs_LPN_ctr'] + contractor_summary['Hrs_CNA_ctr']

# Calculate contractor ratio
contractor_summary['Contractor_Ratio'] = contractor_summary['Total_Contractor_Hours'] / contractor_summary['Total_Staffing_Hours']

# Drop NaN values and facilities with zero staffing
contractor_summary = contractor_summary.dropna().query("Total_Staffing_Hours > 0")

```

```python
contractor_summary = contractor_summary.sort_values(by='Contractor_Ratio', ascending=False)
```

```python
contractor_summary
```

```python
contractor_summary = contractor_summary.sort_values(by='Contractor_Ratio', ascending=False).head(20)
```

```python
# Define color palette
colors = ['#005F73']

# Plot
fig, ax = plt.subplots(figsize=(14, 8))

# Set a custom background color
fig.patch.set_facecolor('#fffbf7')  # Figure background color
ax.set_facecolor('#fffbf7')        # Axis background color

# Create horizontal bar plot
sns.barplot(data=contractor_summary,
            x='Contractor_Ratio',
            y='PROVNAME',
            palette=colors,
            ax=ax)

# Customize title and axis labels
ax.set_title("Top 20 Facilities with Highest Contractor Staffing Ratio (2024Q2)",
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Contractor Staffing Ratio (%)", fontsize=14, labelpad=10)
ax.set_ylabel("Facility Name", fontsize=14, labelpad=10)

# Rotate Y-axis labels for better readability
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right", fontsize=12)

# Add horizontal gridlines
ax.yaxis.grid(True, color="#D3D3D3", linestyle="--", linewidth=0.5, alpha=0.7)
ax.xaxis.grid(False)

# Simplify spines (remove top and right borders)
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

# Annotate bars with contractor ratio values
for index, value in enumerate(contractor_summary['Contractor_Ratio']):
    ax.text(value + 0.01, index, f"{value:.2%}", va='center', fontsize=10, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### 4. Compare RN, LPN, and CNA Staffing
To identify which nursing roles are in highest demand.

```python
# Summing total hours worked by each staff type across all facilities
staffing_by_role = pbj_df.agg({
    'Hrs_RN': 'sum',
    'Hrs_LPN': 'sum',
    'Hrs_CNA': 'sum'
}).reset_index()

# Rename columns for clarity
staffing_by_role.columns = ['Role', 'Total_Hours']

# Convert hours to percentages for better comparison
staffing_by_role['Percentage'] = (staffing_by_role['Total_Hours'] / staffing_by_role['Total_Hours'].sum()) * 100

# Sort roles for visualization
staffing_by_role = staffing_by_role.sort_values(by='Total_Hours', ascending=False)
```

```python
role_colors = ['#005F73', '#0A9396', '#94D2BD']

# Plot Total Staffing by Role
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor('#fffbf7')  # Background color
ax.set_facecolor('#fffbf7')        # Axis background color

sns.barplot(data=staffing_by_role, x='Role', y='Total_Hours', palette=role_colors, ax=ax)

ax.set_title("Total Hours Worked by RN, LPN, and CNA", fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Staff Role", fontsize=14, labelpad=10)
ax.set_ylabel("Total Hours Worked", fontsize=14, labelpad=10)

# Add gridlines
ax.yaxis.grid(True, color="#D3D3D3", linestyle="--", linewidth=0.5, alpha=0.7)
ax.xaxis.grid(False)

# Remove unnecessary spines
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

# Annotate bars with total hours worked
for index, value in enumerate(staffing_by_role['Total_Hours']):
    ax.text(index, value + 0.05 * max(staffing_by_role['Total_Hours']), f"{value:,.0f}",
            ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

1. Certified Nursing Assistants (CNAs) work the most hours  63.25% of all staffing hours in the dataset are attributed to CNAs, indicating that they handle the majority of direct patient care in these facilities.

2. Licensed Practical Nurses (LPNs) come next  They contribute 24.08% of the total staffing hours, showing their critical role in nursing homes but with significantly fewer hours compared to CNAs.

If you're looking at staffing gaps, this suggests that shortages of CNAs would have the greatest impact on patient care, while a shortage of RNs would impact supervision and clinical oversight.

### 5. Contractors vs. Employees in High-Volatility Facilities
This recommendation focuses on facilities with high fluctuations in staffing levels and whether they rely more on contractors vs. full-time employees to manage those variations.

```python
# Convert WorkDate to datetime format if not already
pbj_df['WorkDate'] = pd.to_datetime(pbj_df['WorkDate'], format='%Y%m%d')

# Group by facility and day, summing total staffing hours
staffing_variability = pbj_df.groupby(['PROVNAME', 'WorkDate']).agg({
    'Hrs_RN': 'sum',
    'Hrs_LPN': 'sum',
    'Hrs_CNA': 'sum'
}).reset_index()

# Calculate the daily variation in staffing per facility
staffing_volatility = staffing_variability.groupby('PROVNAME').agg({
    'Hrs_RN': 'std',  # Standard deviation of RN hours worked
    'Hrs_LPN': 'std',  # Standard deviation of LPN hours worked
    'Hrs_CNA': 'std'   # Standard deviation of CNA hours worked
}).reset_index()

# Sum the standard deviations to get an overall volatility score
staffing_volatility['Total_Staffing_Variability'] = staffing_volatility[['Hrs_RN', 'Hrs_LPN', 'Hrs_CNA']].sum(axis=1)

# Sort by highest volatility (greatest fluctuations in staffing hours)
high_variability_facilities = staffing_volatility.sort_values(by='Total_Staffing_Variability', ascending=False)
```

```python
staffing_volatility
```

```python
staffing_mean = pbj_df.groupby(['PROVNAME', 'WorkDate']).agg({
    'Hrs_RN': 'sum',
    'Hrs_LPN': 'sum',
    'Hrs_CNA': 'sum'
}).reset_index()

# Calculate the daily mean in staffing per facility
staffing_mean = staffing_mean.groupby('PROVNAME').agg({
    'Hrs_RN': 'mean',
    'Hrs_LPN': 'mean',
    'Hrs_CNA': 'mean'
}).reset_index()

# Sum the standard deviations to get an overall volatility score
staffing_mean['Total_Staffing_Mean'] = staffing_mean[['Hrs_RN', 'Hrs_LPN', 'Hrs_CNA']].sum(axis=1)

# Sort by highest volatility (greatest fluctuations in staffing hours)
high_mean_facilities = staffing_mean.sort_values(by='Total_Staffing_Mean', ascending=False)
```

```python
staffing_mean
```

```python
staffing_volatility['Total_Staffing_Mean'] = staffing_mean['Total_Staffing_Mean']

# Calculate the absolute difference between Total_Staffing_Variability and Total_Staffing_Mean
staffing_volatility['Variability_Spread'] = abs(staffing_volatility['Total_Staffing_Variability'] - staffing_volatility['Total_Staffing_Mean'])
```

```python
staffing_volatility.head(10)
```

```python
# Sort by highest volatility (greatest fluctuations in staffing hours)
high_variability_facilities = staffing_volatility.sort_values(by='Total_Staffing_Variability', ascending=False)
high_variability_facilities.head(10)
```

```python
# Step 1: Group by facility and sum contractor and employee hours
contractor_summary = pbj_df.groupby('PROVNAME', as_index=False).agg({
    'Hrs_RN_ctr': 'sum',
    'Hrs_LPN_ctr': 'sum',
    'Hrs_CNA_ctr': 'sum',
    'Hrs_RN_emp': 'sum',
    'Hrs_LPN_emp': 'sum',
    'Hrs_CNA_emp': 'sum',
})

# Step 2: Calculate total contractor and employee hours
contractor_summary['Total_Contractor_Hours'] = (
    contractor_summary['Hrs_RN_ctr'] + contractor_summary['Hrs_LPN_ctr'] + contractor_summary['Hrs_CNA_ctr']
)
contractor_summary['Total_Employee_Hours'] = (
    contractor_summary['Hrs_RN_emp'] + contractor_summary['Hrs_LPN_emp'] + contractor_summary['Hrs_CNA_emp']
)

# Step 3: Merge with staffing volatility data (high variability facilities)
staffing_turnover_comparison = pd.merge(high_variability_facilities, contractor_summary, on='PROVNAME', how='left')

# Step 4: Calculate percentage of contractor vs. employee staffing for each facility
staffing_turnover_comparison['Contractor_Percentage'] = (
    staffing_turnover_comparison['Total_Contractor_Hours'] /
    (staffing_turnover_comparison['Total_Contractor_Hours'] + staffing_turnover_comparison['Total_Employee_Hours'])
) * 100

staffing_turnover_comparison['Employee_Percentage'] = 100 - staffing_turnover_comparison['Contractor_Percentage']

# Step 5: Select relevant columns for analysis
staffing_turnover_comparison = staffing_turnover_comparison[
    ['PROVNAME', 'Total_Staffing_Variability', 'Contractor_Percentage', 'Employee_Percentage']
]

# Step 6: Sort by highest staffing variability
staffing_turnover_comparison = staffing_turnover_comparison.sort_values(by='Total_Staffing_Variability', ascending=False).head(20)
```

```python
custom_colors = ['#005F73', '#94D2BD']

fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('#fffbf7')  # Background color
ax.set_facecolor('#fffbf7')        # Axis background color

staffing_turnover_comparison.set_index('PROVNAME')[['Contractor_Percentage', 'Employee_Percentage']].plot(
    kind='bar', stacked=True, color=custom_colors, ax=ax
)

ax.set_title("Contractor vs. Employee Staffing in High-Volatility Facilities",
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Facility Name", fontsize=14, labelpad=10)
ax.set_ylabel("Percentage (%)", fontsize=14, labelpad=10)

plt.legend(title="Staff Type", fontsize=12, title_fontsize=14, loc='upper right')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
